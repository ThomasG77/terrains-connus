```python
import json
import pandas as pd
```

```python
hve = pd.read_csv(
    "https://static.data.gouv.fr/resources/annuaire-des-exploitations-de-haute-valeur-environnementale/20240516-073618/annuaire-des-exploitations-hve-janvier-2024.csv",
    sep=";",
    skiprows=4,
    encoding="ISO-8859-1",
)
hve['code_departement'] = hve['Code Postal'].apply(lambda commune: commune[:2] if commune[:2]!="97" else commune[:3])
hve
```

```python
stats = hve['code_departement'].value_counts().reset_index()
stats.columns = ["departement", "nb_explitations_hve"]
with open("stats_hve.json", "w") as f:
    json.dump(stats.sort_values('departement').to_dict(orient="records"), f)
stats
```
